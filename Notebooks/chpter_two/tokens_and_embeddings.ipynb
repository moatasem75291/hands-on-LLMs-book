# LLM Tokenization

```python
from transformers import AutoTokenizer
```

```python
text = """
English and CAPITALIZATION
ðŸŽµé¸Ÿ
show_tokens False None elif == >= else: two tabs:" " Three tabs:
" "
12.0*50=600
"""
```

```python
colors_list = [
'102;194;165', '252;141;98', '141;160;203',
'231;138;195', '166;216;84', '255;217;47'
]
```

```python
def show_tokens(sentence, tokenizer_name):
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
    tokens_ids = tokenizer(sentence).input_ids
    for idx, t in enumerate(tokens_ids):
        print(
            f'\x1b[0;30;48;2;{colors_list[idx % len(colors_list)]}m'
            + tokenizer.decode(t) +
            '\x1b[0m', end='\n'
            )
```

```python
show_tokens(text, 'google-bert/bert-base-uncased')
```

```python
show_tokens(text, 'google-bert/bert-base-cased')
```

```python
show_tokens(text, 'openai-community/gpt2')
```

```python
show_tokens(text, 'bigcode/starcoder2-15b')
```

```python
show_tokens(text, 'facebook/galactica-120b')
```

```python
show_tokens(text, "microsoft/Phi-3-mini-4k-instruct")
```

## Recommending Songs by Embeddings

```python
import pandas as pd
from urllib import request
```

### Get the playlist dataset

```python
data = request.urlopen("https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt")
```

```python
# Parse the playlist data
lines = data.read().decode("utf-8").split('\n')[2:]
```

```python
# Remove playlists with only one song
playlists = [s.rstrip().split() for s in lines if len(s.split()) > 1]
```

____

### Load song metadata

```python
songs_file = request.urlopen(
    'https://storage.googleapis.com/maps-premium/dataset/yes_complete/song_hash.txt'
    )
```

```python
songs_file = songs_file.read().decode("utf-8").split('\n')
songs = [s.rstrip().split('\t') for s in songs_file]
```

```python
songs_df = pd.DataFrame(songs, columns=['song_id', 'song_name', 'artist'])
```

```python
songs_df = songs_df.set_index('song_id')
songs_df
```

```python
print('Playlist #1:\n ', playlists[0])
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    playlists,
    vector_size=32,
    window=20,
    negative=50,
    min_count=1,
    workers=4
)
```
